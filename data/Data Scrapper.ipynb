```python
import requests
from bs4 import BeautifulSoup

from selenium import webdriver
import time

import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.DataFrame(columns=['title', 'intro', 'article'])
```

```python
all_links = []
for i in range(0, 500, 7):
    page = requests.get('http://texty.org.ua/mod/article/?offset=' + str(i))
    soup = BeautifulSoup(page.content, 'html.parser')
    weblinks = soup.find_all('article')
    for link in weblinks:
        url = link.find_all('a')[0]
        all_links.append(url.get('href'))
```

```python
count = 0

for link in all_links:
    count += 1
    # store the text for each article
    paragraphtext = []    
    # get url
    url = link
    # get page text
    page = requests.get(url)
    # parse with BFS
    soup = BeautifulSoup(page.text, 'html.parser')    
    
    # get article title
    atitle = soup.find(class_="art-title")
    title = atitle.get_text() 
    print(str(count) + ' - ' + title)
    
    # get article intro
    aintro = soup.find(class_="article-intro").p
    intro = aintro.get_text()
    
    # get article content
    acontent = soup.find(class_="article-content")
    content = acontent.find_all('p', recursive=False)   
    for paragraph in content:
        # get the text only
        text = paragraph.get_text()
        paragraphtext.append(text)        
    
    article = ' '.join(paragraphtext)
    df = df.append({'title': title, 'intro': intro, 'article': article}, ignore_index=True)
```
